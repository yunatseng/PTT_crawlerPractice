```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://finfo.tw/hottest'
query = '?period=month'
```

```python
response = requests.get(base_url)
soup = BeautifulSoup(response.text,"html.parser")
sections = soup.select('.product-infos > h3 > a')
sections
```

```python
sections[0].attrs['href']
```

```python
json_week = {
    'name': [i.text for i in sections],
    'url': ['https://finfo.tw'+ i.attrs['href'] for i in sections],
}
json_week
```

```python
response_month = requests.get(base_url + query)
soup = BeautifulSoup(response.text,"html.parser")
sections_month = soup.select('.product-infos > h3 > a')
sections_month
```

```python
json_month = {
    'name': [i.text for i in sections_month],
    'url': ['https://finfo.tw'+ i.attrs['href'] for i in sections_month],
}
json_month
```

```python
json_all={
    'month':json_month,
    'week':json_week
}
json_all
```
