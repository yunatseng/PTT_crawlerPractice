```python
import requests
from bs4 import BeautifulSoup
```

```python
my_headers = {'cookie': 'over18=1;'}
base_url = 'https://www.ptt.cc'
```

```python
try:
    URL = base_url+'/bbs/Gossiping/search?page=6&q=%E5%B9%B9%E4%BD%A0%E5%A8%98%EF%BC%8C%E5%B1%8F%E6%9D%B1%E6%8C%96%E7%9C%BC%E6%A1%88'
    response = requests.get(URL, headers = my_headers)
    soup = BeautifulSoup(response.text,"html.parser")
    metaSection = soup.select('div.r-list-container') [0]
    sections = metaSection.select('div.r-ent')
    target_url = []
    
    #先試試看爬文章標題作者日期
    for section in sections:
        title = section.select('div.title')[0].text
        author = section.select('div.author')[0].text
        date = section.select('div.date') [0].text
        title = title.strip()
        href = str(section.select('div.title a')[0]).split('"')[1]
        
        print(title)
        print(author)
        print(date)
        print(href)
        target_url.append(href)
    print(target_url)
    
except Exception as e:
    print(e)
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)

try:
    URL = base_url+'/bbs/Gossiping/search?page=7&q=%E5%B9%B9%E4%BD%A0%E5%A8%98%EF%BC%8C%E5%B1%8F%E6%9D%B1%E6%8C%96%E7%9C%BC%E6%A1%88'
    response = requests.get(URL, headers = my_headers)
    soup = BeautifulSoup(response.text,"html.parser")
    metaSection = soup.select('div.r-list-container') [0]
    sections = metaSection.select('div.r-ent')

    #再試試看爬內文
    ptt_content_list = []
    for section in sections:
        href = str(section.select('div.title a')[0]).split('"')[1]
        ARTICLE_URL = base_url + href
        r = requests.get(ARTICLE_URL, headers = my_headers)
        soup = BeautifulSoup(r.text,"html.parser")       
        meta_data = soup.select('span.article-meta-value')
        author = meta_data[0].text
        date = meta_data[3].text       
        main_content = soup.find(id='main-content').text.split('--')[0]
        separator = '\n'
        main_text = main_content.split(separator)[1:]
        final_text = separator.join(main_text)        
        ptt_content = {'author': author,'date': date,'final_text': final_text.replace('\n',' ')}       
        ptt_content_list.append(ptt_content)
        
#         print(ptt_content)
#         print("=============")
#         print(author)
#         print(date)
#         print(final_text)
#         print("=============")
    pp.pprint(ptt_content_list)
    
except Exception as e:
    print(e)
```
